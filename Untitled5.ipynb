**Word Game using Python Project 2**

```python
import random
```

```python
words=["Python","Java","C language","Html","CSS","Javascript","React js","Node js","Mango DB","C++"]
```

```python
random.choice(words)
```

```python
guessed_word = random.choice(words)
guessed_word
```

```python
hint=guessed_word[2] + guessed_word[3]
hint
```

```python
store_g_l = []
try_p = 5
a = input('Enter your name')
print('Welcome to word game',a)
print('We are going to give you 5 chances for the guessing word')
```

```python
for guess in range(try_p):
  while True:
    letter = input("Guess the letter:")

    if len(letter) == 1:
      break
    else:
      print("Oops! Please guess a letter!")

  if letter in guessed_word:
    print("Good guess!")
    store_g_l.append(letter)
  else:
    print("Bad guess!")

  if guess == 3:
    print()
    clue_request = input("Would you like a clue? (yes/no): ")
    if clue_request.lower() == "yes":
      print("clue:", hint)
    else:
      print("You are very confident!")

print()
print("The word you guess is correct:",len(store_g_l))
print("These letters are:", store_g_l)

word_guess = input("Guess the word:")
if word_guess.lower() == guessed_word.lower():
  print("Congratulations!!! you won the game")
else:
  print("Sorry!!! you lost the game")

print("The correct word is:",guessed_word)
print("Better luck next time")


```

**Calculator using Python Project 3**

```python
def add(x,y):
  return(x + y)

def sub(x,y):
  return(x - y)

def mul(x,y):
  return(x * y)

def div(x,y):
  return(x / y)

def mod(x,y):
  return(x % y)

def exp(x,y):
  return(x ** y)

print("Select operation:")
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")
print("5.Modulus")
print("6.Exponent")

while True:
  choice = input("Enter choice(1/2/3/4/5/6)")
  if choice in ('1','2','3','4','5','6'):
    num1 = float(input("Enter first number:"))
    num2 = float(input("Enter second number:"))

    if choice == '1':
      print(num1,"+",num2,"=",add(num1,num2))

    elif choice == '2':
      print(num1,"-",num2,"=",sub(num1,num2))

    elif choice == '3':
      print(num1,"*",num2,"=",mul(num1,num2))

    elif choice == '4':
      print(num1,"/",num2,"=",div(num1,num2))

    elif choice == '5':
      print(num1,"%",num2,"=",mod(num1,num2))

    elif choice == '6':
      print(num1,"**",num2,"=",exp(num1,num2))
    break

  else:
    print("Invalid Input")
```

**Uber Data Analysis Project 4**

```python
from ast import increment_lineno
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import gc
import os
import sys
%matplotlib inline
```

```python
def reduce_memory_usage(df, verbose=True):
  numerics = ['int16','int32','int64','float16','float32','float64']
  start_mem = df.memory_usage().sum()/1024**2
  for col in df.columns:
    col_type = df[col].dtypes
    if col_type in numerics:
      c_min = df[col].min()
      c_max = df[col].max()
      if str(col_type)[:3] == 'int':
        if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
          df[col] = df[col].astype(np.int8)
        elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
          df[col] = df[col].astype(np.int16)
        elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
          df[col] = df[col].astype(np.int32)
        elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
          df[col] = df[col].astype(np.int64)
      else:
        if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
          df[col] = df[col].astype(np.float32)
        elif c_min > np.finfo(np.float64).min and c_max < np.finfo(np.float64).max:
          df[col] = df[col].astype(np.float64)
        else:
          df[col] = df[col].astype(np.float64)
  end_mem = df.memory_usage().sum()/1024**2
  if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100*(start_mem-end_mem)/start_mem))
  return df
```

```python
cab_data = pd.read_csv('/content/cab_rides.csv')
cab_data = reduce_memory_usage(cab_data)
weather_data = pd.read_csv('/content/weather_forecast_data.csv')
weather_data = reduce_memory_usage(weather_data)
```
